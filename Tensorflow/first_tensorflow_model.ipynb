```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import Sequential 
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam,SGD
from tensorflow.keras.losses import mean_squared_error,binary_crossentropy
from tensorflow import random_uniform_initializer
```

```python
data=np.load('Tf_intro.npz')
data
```

```python
input_size=2
output_size=1
model=Sequential([Dense(units=1,kernel_initializer=random_uniform_initializer(minval=-0.1,maxval=0.1),
                        bias_initializer=random_uniform_initializer(minval=-0.1,maxval=0.1))])
custom_optimizer=SGD(learning_rate=0.02)
model.compile(optimizer=custom_optimizer,loss='mean_squared_error')
model.fit(data['inputs'],data['targets'],epochs=100,verbose=2)
```

```python
#Extract weights and bias
weights=model.layers[0].get_weights()[0]
bias=model.layers[0].get_weights()[1]
print("weights",weights)
print("bias",bias)
```

```python
model.predict_on_batch(data['inputs']).round(1)
```
