```python
from tensorflow.keras.preprocessing.image import load_img,ImageDataGenerator
from tensorflow.keras.preprocessing import image
import random
import os
```

```python
selection=random.choice(os.listdir(r'D:\machine_learning\Tensorflow\car_model_learning\Datasets\Test\lamborghini'))
```

```python
image=load_img('D:\\machine_learning\\Tensorflow\\car_model_learning\\Datasets\\Test\\lamborghini\\'+selection)
```

```python
image
```

```python
datagen=ImageDataGenerator(rotation_range=40,shear_range=0.2,
                          height_shift_range=0.3,width_shift_range=0.3,zoom_range=0.5,horizontal_flip=True,fill_mode="nearest")
```

```python
from tensorflow.keras.preprocessing.image import img_to_array
x=img_to_array(image)
x.shape[0]
```

```python
x=x.reshape((1,)+x.shape)
x
```

```python
i=0
for images in datagen.flow(x,save_to_dir="datagen",save_prefix="lambo",save_format="jpeg"):
    i+=1
    if i>20:
        break
```
